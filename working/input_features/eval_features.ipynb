```python
# featurizer 1: Circular Fingerprint (ECFP)

df_chembl_ecoli_wmol = featurize.molstr_to_Mol(df_chembl_ecoli, strcolumnID='smiles')
chembl_w_CF = featurize.add_features(df_chembl_ecoli_wmol)

df_zhu_rat_wmol = featurize.molstr_to_Mol(df_zhu_rat, strcolumnID='smiles')
zhu_w_CF = featurize.add_features(df_zhu_rat_wmol)

#error
df_fillipo_wmol = featurize.molstr_to_Mol(df_fillipo, strcolumnID='smiles')
fillipo_w_CF = featurize.add_features(df_fillipo_wmol)


```

```python
dataset_chembl_CF = dataprep.convert_to_dataset(chembl_w_CF, 
                             X_col= 'CircularFingerprint', 
                             y_col= 'MIC', 
                             w_col= None,
                             id_col=None)
dataset_zhu_CF = dataprep.convert_to_dataset(zhu_w_CF, 
                             X_col= 'CircularFingerprint', 
                             y_col= 'LD50', 
                             w_col= None,
                             id_col=None)
```

```python
# transformed_data = dataprep.data_transformation(dataset, 
#                         transformations= ['NormalizationTransformer'], 
#                         to_transform= ['X'])
```

```python
train_chembl_CF, test_chembl_CF = dataprep.data_splitting(dataset_chembl_CF,  
                   split_type = 'train_test_split', 
                   k = None, 
                   frac_train = 0.7,  
                   log_every_n= None)
```

```python
ToxModel.help()
```

```python
model_chembl_CF = ToxModel('LASSO')
```

```python
model_chembl_CF.fit(train_chembl_CF)
```

```python
model_chembl_CF.evaluate(test_chembl_CF, metrics=['r2_score', 'mean_squared_error'])
```

```python
model_chembl_CF.visualize('pair_predict', test_chembl_CF, untransform=False)
```
