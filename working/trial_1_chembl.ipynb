```python
import cytoxnet
from cytoxnet.dataprep import io
from cytoxnet.dataprep import featurize
import dataprep
import deepchem as dc
import numpy as np
from cytoxnet.models.models import ToxModel
```

```python
df_orig = io.load_data('ChemBL_ATCC25922_MIC_2.csv')
```

```python
data = io.load_data('ChemBL_ATCC25922_MIC_2.csv', col_id='Molecule ChEMBL ID')
```

```python
data
```

```python
data = data[data['Smiles'].notna()]

data = data[data['Standard Value'].notna()]

data_w_mol = featurize.molstr_to_Mol(data, strcolumnID='Smiles')

data_feat = featurize.add_features(data_w_mol)
data_feat = data_feat.reset_index(drop=True)
```

```python
data_feat
```

```python
def convert_to_dataset(dataframe, 
                       X_col: str = 'X', 
                       y_col: str = 'y', 
                       w_col: str = None ,
                       id_col: str = None, 
                       return_csv = False):
    """
    Converts dataframe into a deepchem dataset object.
    
    Parameters
    ----------
    - dataframe: featurized dataframe
    - X_col: (str or List[str]) name(s) of the column(s) containing the X array.
    - y_col: (str or List[str]) name(s) of the column(s) containing the y array.
    - w_col: (str or List[str]) name(s) of the column(s) containing the w array.
    - id_col: (str) name of the column containing the ids.
    - return_csv: (True/False) whether a viewable csv of the data will be returned with the dataset object.

    Returns
    -------
    - deepchem dataset
    - (conditional) csv of dataframe
    """
    # define x
    X = dataframe[X_col].values
    if type(X[0][0]) is np.ndarray:
        print('yes')
        X = np.stack(X).reshape(len(dataframe), -1)
    else:
        X = np.stack(X).reshape(len(dataframe))
        
    # define y
    y = dataframe[y_col].values
    
    # define weight
    if w_col is not None:
        w = dataframe[w_col]
    else:
        w = None
    
    # define ids
    if id_col is not None:
        ids = dataframe[id_col]
    else:
        ids = None
    
    # create deepchem dataset object
    dataset = dc.data.NumpyDataset(X, y, w, ids)
    
    # return dataset and csv of current dataframe if return_csv equals True
    # return only dataset if return_csv equals False or unspecified
    if return_csv is True:
        csv = dataframe.to_csv(index = True, header=True)
        return dataset, csv
    else:
        return dataset
```

```python
dataset = convert_to_dataset(data_feat, 
                             X_col= 'CircularFingerprint', 
                             y_col= 'Standard Value', 
                             w_col= None ,
                             id_col = None, 
                             return_csv = False)


```

```python
transformed_data = dataprep.data_transformation(dataset, 
                        transformations= ['NormalizationTransformer'], 
                        to_transform= ['X'])
```

```python
train, test = dataprep.data_splitting(transformed_data,  
                   split_type = 'train_test_split', 
                   k = None, 
                   frac_train = 0.7, 
                   frac_valid = None, 
                   frac_test = None, 
                   log_every_n= None)
```

```python
ToxModel.help()
```

```python
model = ToxModel('LASSO')
```

```python
model.fit(train)
```

```python
model.evaluate(test, metrics=['r2_score', 'mean_squared_error'])
```

```python
model.visualize('pair_predict', test, untransform=False)
```
